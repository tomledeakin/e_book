4.
Kiểm tra tất cả các giá trị trong tuple bất kì (n phần tử) trùng nhau hay không:
VD:(10, 10, 10, 10) -> True
(10, 10, 2) -> False


```python
tuple = (10, 10, 10, 2)
count = 0
bool = "False"
for index in range(len(tuple)):
  if tuple[0] == tuple[index]:
    count += 1
if count == 0:
  bool = "False"
else:
  bool = "True"
print(bool)
```

5.
Cho 2 set,
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
a. Tìm set các phần tử có trong set1 hoặc set2 ( phần tử có cả trong set1 và set2 vẫn thỏa mãn).
b. Tìm set các phần tử có trong cả set1 và set2
c. Tìm set các phần tử có trong set1 nhưng không trong set2


```python
# Cho 2 set
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}

# a. Tìm set các phần tử có trong set1 hoặc set2
union_set = set1 | set2
print("a. Union Set:", union_set)

# b. Tìm set các phần tử có trong cả set1 và set2
intersection_set = set1 & set2
print("b. Intersection Set:", intersection_set)

# c. Tìm set các phần tử có trong set1 nhưng không trong set2
difference_set = set1 - set2
print("c. Difference Set:", difference_set)
```

8.
Cho 1 list các số:
a. tạo dict với key là số, value là số lần xuất hiện số đấy.
num = [1, 2, 3, 1, 2, 1, 1, 4] => dict = {1:4, 2:2, 3:1,4:1}
b. in ra số mà có lần xuất hiện nhiều nhất và ít nhất trong dãy ( nhiều nhất: 1, ít nhất: 3,4)
c. in ra giá trị trung bình value của dict, (4+2+1+1)/4


```python
num = [1, 2, 3, 1, 2, 1, 1, 4]

# a. Tạo dict với key là số, value là số lần xuất hiện số đó
num_dict = {}
for number in num:
    if number in num_dict:
        num_dict[number] += 1
    else:
        num_dict[number] = 1

print("a. Dictionary:", num_dict)

# b. In ra số xuất hiện nhiều nhất và ít nhất trong dãy
max_occur = max(num_dict, key=num_dict.get)
min_occur = min(num_dict, key=num_dict.get)
print("b. Số xuất hiện nhiều nhất:", max_occur, "Số xuất hiện ít nhất:", min_occur)

# c. In ra giá trị trung bình value của dict
average_value = sum(num_dict.values()) / len(num_dict)
print("c. Giá trị trung bình của value:", average_value)
```

9.
Cho 1 dict điểm học sinh:
scores = {'Tuan': 8, 'Minh': 7, 'Hùng': 9, 'Hiếu': 10}
a. Cho người dùng nhập tên học sinh nếu học sinh có trong dict thì in ra điểm, nếu không thì in ra học sinh không có trong lớp.
b. Cho người dùng nhập tên và điểm của 1 học sinh, thêm thông tin vào dict
c. In ra các bạn có điểm cao nhất / thấp nhất trong lớp.


```python
scores = {'Tuan': 8, 'Minh': 7, 'Hùng': 9, 'Hiếu': 10}

student_name_a = input("a. Nhập tên học sinh: ")
if student_name_a in scores:
    print(f"Điểm của {student_name_a}: {scores[student_name_a]}")
else:
    print(f"{student_name_a} không có trong lớp.")

student_name_b = input("b. Nhập tên học sinh: ")
student_score_b = int(input("Nhập điểm học sinh: "))
scores[student_name_b] = student_score_b
print(f"Thông tin của {student_name_b} đã được thêm vào dict: {scores}")

max_score_student = max(scores, key=scores.get)
min_score_student = min(scores, key=scores.get)
max_score = scores[max_score_student]
min_score = scores[min_score_student]

print(f"c. Học sinh có điểm cao nhất là {max_score_student} với {max_score}")
print(f"   Học sinh có điểm thấp nhất là {min_score_student} với {min_score}")

# EM KHÔNG BIẾT VÌ SAO NÓ CỨ BÁO LỖI KHÔNG LIÊN QUAN Ạ NÊN EM KHÔNG CHECK ĐƯỢC LỖI ĐÚNG HAY SAI, MÀ VỚI CẢ SỬ DỤNG FUNCTION LIST() ĐỂ CHUYỂN TỪ SET VỀ LIST
```

10.
Cho người dùng nhập tên sản phẩm và số lượng sản phẩm cho đến khi người dùng nhập tên là "end" thì kết thúc vòng lặp. Lưu dữ liệu dạng dict: tên_sản phẩm: số lượng. Nếu sản phẩm có rồi thì cộng dồn số lượng sản phẩm, nếu chưa có thì thêm mới vào dict.
a. Cho người dùng nhập tên sản phẩm in ra số lượng, nếu tên không có báo không có sản phẩm này.
b. In ra các sản phẩm có số lượng lớn nhất.


```python
product_dict = {}

# a. Cho người dùng nhập tên sản phẩm và in ra số lượng
while True:
    product_name = input("Nhập tên sản phẩm (hoặc 'end' để kết thúc): ").lower()

    if product_name == 'end':
        break  # Kết thúc vòng lặp nếu nhập 'end'

    quantity = int(input("Nhập số lượng sản phẩm: "))

    # Kiểm tra xem sản phẩm đã có trong dict hay chưa
    if product_name in product_dict:
        product_dict[product_name] += quantity  # Cộng dồn số lượng nếu sản phẩm đã có
    else:
        product_dict[product_name] = quantity  # Thêm mới sản phẩm vào dict nếu chưa có

# b. In ra các sản phẩm có số lượng lớn nhất
if product_dict:
    max_quantity = max(product_dict.values())
    max_quantity_products = [product for product, quantity in product_dict.items() if quantity == max_quantity]

    print("\nSố lượng lớn nhất là:", max_quantity)
    print("Các sản phẩm có số lượng lớn nhất:", max_quantity_products)
else:
    print("Danh sách sản phẩm trống.")
```
